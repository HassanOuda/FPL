```python
#Import models from scikit learn module:
from sklearn.linear_model import LogisticRegression
from sklearn.cross_validation import KFold   #For K-fold cross validation
from sklearn.ensemble import RandomForestClassifier
from sklearn.tree import DecisionTreeClassifier, export_graphviz
from sklearn import metrics
import numpy as np

#Generic function for making a classification model and accessing performance:
def classification_model(model, data, predictors, outcome):
  #Fit the model:
  model.fit(data[predictors],data[outcome])
  
  #Make predictions on training set:
  predictions = model.predict(data[predictors])
  
  #Print accuracy
  accuracy = metrics.accuracy_score(predictions,data[outcome])
  print ("Accuracy : %s" % "{0:.3%}".format(accuracy))

  #Perform k-fold cross-validation with 5 folds
  kf = KFold(data.shape[0],n_folds=5)
  error = []
  for train, test in kf:
    # Filter training data
    train_predictors = (data[predictors].iloc[train,:])
    
    # The target we're using to train the algorithm.
    train_target = data[outcome].iloc[train]
    
    # Training the algorithm using the predictors and target.
    model.fit(train_predictors, train_target)
    
    #Record error from each cross-validation run
    error.append(model.score(data[predictors].iloc[test,:], data[outcome].iloc[test]))
 
  print ("Cross-Validation Score : %s" % "{0:.3%}".format(np.mean(error)))

  #Fit the model again so that it can be refered outside the function:
  model.fit(data[predictors],data[outcome]) 
```

```python
outcome_var = 'prev_total_points'
model = LogisticRegression()
predictor_var = ['bps','chance_of_playing_this_round','value_form','influence','creativity','threat','minutes','prev_was_home',
                'prev_assists','prev_goals_scored','element_type']
classification_model(model, playerInfo,predictor_var,outcome_var)
```

```python
model = DecisionTreeClassifier()
predictor_var = ['prev_attempted_passes', 
       'prev_bonus', 'prev_bps', 'prev_clean_sheets','prev_creativity', 'prev_influence','prev_minutes','prev_yellow_cards',
       'prev_clearances_blocks_interceptions', 'prev_completed_passes','prev_dribbles','prev_assists',
       'prev_fouls',
        'prev_opponent_team', 
       'prev_tackles',
       'prev_value','element_type','prev_saves','prev_clean_sheets']
classification_model(model, playerInfo,predictor_var,outcome_var)
#Create a series with feature importances:
featimp = pd.Series(model.feature_importances_, index=predictor_var).sort_values(ascending=False)
print (featimp)
```

```python
model = RandomForestClassifier(n_estimators=25, min_samples_split=25, max_depth=7, max_features=1)
predictor_var =   ['prev_minutes','prev_creativity','prev_influence','prev_clearances_blocks_interceptions',
'prev_recoveries','prev_attempted_passes','prev_fouls','prev_tackled','prev_clean_sheets',
'prev_bps','prev_key_passes','prev_selected','prev_completed_passes','prev_value',
'prev_big_chances_missed','prev_target_missed','element_type','prev_saves','prev_penalties_saved','prev_clean_sheets']
classification_model(model, playerInfo,predictor_var,outcome_var)
#Create a series with feature importances:
featimp = pd.Series(model.feature_importances_, index=predictor_var).sort_values(ascending=False)
print (featimp)
```
